# Informe de resultados

## Análisis de las variables numéricas

### Análisis de las variables relacionadas con los vuelos


Las variables relacionadas con la actividad de vuelos muestran un patrón común: una parte importante de los registros mensuales presenta valores bajos o nulos, mientras que una minoría concentra niveles de actividad más elevados.

En el caso de `flights_booked`, la diferencia entre la media y la mediana sugiere una distribución asimétrica hacia la derecha que se confirma en el gráfico de barras y de bigotes. Esto indica que la mayoría de los registros mensuales concentra pocos vuelos reservados, mientras que algunos registros con valores más altos elevan la media. Esta interpretación se refuerza con la moda, ya que el valor más frecuente es 0. Es decir, lo más habitual es que en un mes concreto un cliente no haya reservado vuelos.

La variable `flights_with_companions` presenta un comportamiento todavía más concentrado en valores bajos. La media y la mediana están muy próximas, y la mediana coincide con la moda, ambas con valor 0. Esto significa que al menos la mitad de los registros mensuales no incluye vuelos con acompañantes. Aunque el valor máximo alcanza los 11 vuelos con acompañante, el boxplot muestra que estos casos son minoritarios y atípicos frente al comportamiento general de la variable.

Por su parte, `total_flights` vuelve a mostrar una distribución asimétrica, con una media superior a la mediana. Esto indica que algunos registros con un número elevado de vuelos desplazan la media hacia arriba. Además, la desviación estándar, la varianza y el coeficiente de variación reflejan una dispersión elevada, coherente con una variable en la que conviven muchos registros sin actividad con otros que presentan un volumen de vuelos notablemente superior.

En conjunto, el análisis muestra que la actividad mensual de los clientes dentro del programa de fidelización no se distribuye de forma homogénea. La mayoría de los registros se concentra en niveles bajos de actividad, mientras que una minoría de clientes o meses acumula un mayor número de vuelos. Esta diferencia entre registros activos e inactivos es relevante para interpretar el comportamiento de los clientes y evitar que los valores más altos distorsionen la lectura del promedio.

### Análisis de distancia y puntos

Las variables `distance`, `points_accumulated`, `points_redeemed` y `dollar_cost_points_redeemed` presentan un comportamiento similar al observado en las variables de vuelos. En todas ellas se aprecia una fuerte concentración de registros con valor 0, lo que está relacionado con la elevada presencia de meses sin actividad registrada por parte de los clientes.

En el caso de `distance`, la mayoría de los registros mensuales acumula poca o ninguna distancia, mientras que una minoría presenta distancias elevadas que desplazan la media hacia arriba. Dado que los desplazamientos en avión pueden implicar distancias grandes incluso con pocos vuelos, la variable muestra una dispersión elevada.

La variable `points_accumulated` sigue una lógica similar, ya que la acumulación de puntos está estrechamente vinculada a la actividad de vuelo. La presencia de muchos registros sin vuelos explica que la moda vuelva a ser 0 y que la distribución no sea simétrica. Los valores más altos de puntos acumulados corresponden a registros con mayor actividad, lo que incrementa la distancia entre las medidas centrales y eleva la dispersión.

Además, el análisis de correlación entre `distance` y `points_accumulated` muestra una relación positiva muy fuerte, con un coeficiente de correlación de Pearson de 0,99. Esto indica que ambas variables evolucionan prácticamente en la misma dirección: a medida que aumenta la distancia recorrida, también aumentan los puntos acumulados. Este resultado es coherente con el funcionamiento habitual de un programa de fidelización, en el que los puntos suelen estar vinculados a la distancia o actividad de vuelo.

En conjunto, estas variables reflejan un patrón claro de polarización en la actividad mensual: muchos registros sin actividad y una minoría de registros con valores elevados de distancia, puntos acumulados o puntos canjeados. Por este motivo, la media debe interpretarse con cautela, ya que puede estar influida por valores altos poco frecuentes.

### Análisis de Salary y CLV



La variable `salary` presenta una distribución asimétrica hacia la derecha. El 50% central de los salarios se sitúa entre 63.991 y 82.940, mientras que los valores considerados dentro del rango habitual llegan aproximadamente hasta 111.106. A partir de ese umbral aparecen numerosos valores atípicos hacia la derecha, con un máximo superior a 400.000, lo que amplía notablemente el rango de la variable y contribuye a aumentar la dispersión.

También se observan algunos valores atípicos en el extremo inferior de la distribución, aunque son menos numerosos y menos relevantes que los valores altos. En conjunto, el salario muestra una dispersión elevada, reflejada tanto en el rango como en el coeficiente de variación.

No obstante, la interpretación de esta variable debe hacerse con cautela. Tras la imputación de los valores nulos, la media y la mediana coinciden, y la moda se concentra en el valor central imputado. Esto indica que el método utilizado ha generado una acumulación de valores en torno a un mismo salario estimado, reduciendo parcialmente la variabilidad real de la variable.

Por su parte, la variable `clv` también presenta una distribución asimétrica hacia la derecha. La mayoría de los clientes se concentra en valores relativamente bajos de CLV, mientras que una minoría alcanza valores mucho más elevados. Esta presencia de valores extremos desplaza la media hacia arriba, por lo que la mediana, situada en 5.780,18, resulta una medida más representativa del valor típico de los clientes.

Al analizar conjuntamente `salary` y `clv`, el gráfico de dispersión no muestra una relación lineal clara entre ambas variables. La mayor parte de los clientes se concentra en salarios entre 50.000 y 100.000, pero dentro de ese rango se observan valores de CLV muy diferentes. Además, los salarios más altos no se asocian necesariamente con un mayor CLV.

Esta interpretación es coherente con el coeficiente de correlación de Pearson, que se sitúa en torno a -0.02, un valor muy próximo a 0. Por tanto, a nivel descriptivo, no se observa una relación lineal significativa entre el salario de los clientes y su valor de vida estimado dentro del programa.

### Análisis de variables categóricas

El análisis de las variables categóricas permite conocer mejor el perfil general de los clientes del programa de fidelización. Para evitar duplicidades, estas variables se analizaron a nivel cliente, ya que representan características individuales y no actividad mensual.

En primer lugar, la distribución por género se encuentra prácticamente equilibrada. El número de mujeres y hombres es muy similar, con 8.410 mujeres y 8.327 hombres, por lo que no se observa una diferencia relevante en la composición de la base de clientes según esta variable.

En cuanto al nivel educativo, la categoría más frecuente es `Bachelor`, con una diferencia muy marcada frente al resto de niveles. Le sigue `College`, mientras que `High School or Below`, `Doctor` y `Master` tienen una representación mucho menor. Esto indica que la base de clientes se concentra principalmente en niveles educativos medios y superiores, especialmente en la categoría `Bachelor`.

Respecto al estado civil, la mayoría de los clientes están casados, seguidos por clientes solteros y, en menor medida, divorciados. Esta distribución muestra una concentración clara en la categoría `Married`, que representa el grupo principal dentro de la variable `marital_status`.

En relación con el tipo de tarjeta de fidelidad, la categoría más frecuente es `Star`, seguida de `Nova` y, finalmente, `Aurora`. Esto sugiere que la mayor parte de los clientes se encuentra en el nivel de tarjeta más común del programa, mientras que las categorías superiores o menos frecuentes concentran un menor número de usuarios.

La variable `enrollment_type` muestra una distribución muy desigual. La gran mayoría de clientes pertenece al tipo de inscripción `Standard`, mientras que la promoción de 2018 representa una proporción muy reducida del total. Esto indica que la captación de clientes se produce principalmente a través de la modalidad estándar.

Por último, la distribución por provincia está claramente concentrada en unas pocas regiones. Ontario, British Columbia y Quebec agrupan la mayor parte de los clientes, mientras que el resto de provincias tienen una representación mucho menor. Esta concentración territorial puede estar relacionada con el tamaño de la población, la presencia comercial de la compañía o la mayor actividad del programa en determinadas zonas.

En conjunto, las variables categóricas muestran una base de clientes equilibrada por género, pero concentrada en determinados perfiles educativos, estados civiles, tipos de inscripción y provincias. Destacan especialmente la alta presencia de clientes con nivel educativo `Bachelor`, clientes casados, usuarios con tarjeta `Star` e inscripciones estándar.

### Salario promedio por nivel educativo

El análisis del salario promedio por nivel educativo muestra diferencias claras entre los distintos grupos. Los clientes con nivel educativo `Doctor` presentan el salario promedio más alto, con 178.609, seguidos por los clientes con `Master`, cuyo salario promedio se sitúa en 103.758.

En los niveles intermedios, `College` presenta un salario promedio de 79.430, mientras que `Bachelor` se sitúa ligeramente por debajo, con 72.658. Por último, el grupo `High School or Below` registra el salario promedio más bajo, con 61.237.

En conjunto, el gráfico muestra una tendencia general en la que los niveles educativos más altos se asocian con salarios promedio superiores, especialmente en los casos de `Master` y `Doctor`. No obstante, esta interpretación debe hacerse con cautela, ya que la variable `salary` fue imputada y presenta valores atípicos que pueden influir en la media.

## ¿Qué variables pueden influir en la reserva de un mayor número de vuelos?

### Vuelos reservados por nivel educativo


El nivel educativo no parece generar diferencias relevantes en el total de vuelos reservados. Las medias y medianas por grupo son muy similares, y los boxplots muestran distribuciones parecidas entre los distintos niveles educativos. Por tanto, a nivel descriptivo, la educación no parece ser una variable claramente asociada a una mayor reserva de vuelos.

### Relación entre salario y vuelos reservados

El análisis de la relación entre `salary` y el total de vuelos reservados por cliente no muestra una asociación clara entre ambas variables. El coeficiente de correlación es de 0,01, un valor prácticamente igual a 0, lo que indica ausencia de relación lineal significativa.

En el gráfico se observa que la mayor parte de los clientes se concentra en salarios entre 50.000 y 100.000, pero dentro de ese rango hay una gran variedad de comportamientos: clientes con pocos vuelos, clientes con niveles medios de actividad y clientes con muchos vuelos reservados. Además, los salarios más altos no se asocian necesariamente con un mayor número de vuelos, como en un primer momento se podía esperar.

Por tanto, a nivel descriptivo, el salario no parece ser una variable determinante para explicar la reserva de un mayor número de vuelos dentro del programa de fidelización.

En resumen, aunque inicialmente podría esperarse que los clientes con mayor salario reservaran más vuelos, los datos no muestran esa relación. La actividad de reserva parece depender de otros factores distintos al nivel salarial.

### Vuelos reservados por mes

El gráfico muestra la cantidad total de vuelos reservados por mes, agregando los datos disponibles de 2017 y 2018. Esta visualización permite observar un claro patrón estacional en la reserva de vuelos.

Los meses con mayor volumen de reservas son julio, junio y agosto, lo que coincide con el periodo vacacional de verano. También destaca diciembre, que presenta un nivel elevado de reservas y puede estar asociado al periodo navideño.

Por el contrario, enero y febrero registran los valores más bajos del año, lo que sugiere una menor actividad tras el periodo vacacional de invierno.

En conjunto, los datos apuntan a que la demanda de vuelos reservados aumenta especialmente durante los meses asociados a vacaciones, principalmente en verano y, en menor medida, en diciembre.

Este resultado sugiere que la estacionalidad puede ser una variable relevante para explicar el volumen de vuelos reservados. Más que determinadas características individuales del cliente, como el salario, el momento del año parece tener una relación más clara con el aumento de las reservas.
